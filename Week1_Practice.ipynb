```python
#Write FizzBuzz (1–30) three ways: (a) using if/elif/else, (b) using a dictionary mapping, (c) using a one-liner with a ternary expression.
#Compare readability. Which style do you see most in real Python code?

# a) using if/elif/else

for i in range(1, 31):
  if i%3 == 0 and i%5 == 0:
    print("FizzBuzz")
  elif i%3 == 0:
    print("Fizz")
  elif i%5 == 0:
    print("FizzBuzz")
  else:
    print(i)

# b)  using a dictionary mapping
mapping = {
    (True, True) : "FizzBuzz",
    (True, False) : "Fizz",
    (False, True) : "Buzz",
    (False, False) : None
  }

for i in range(1, 31):
  key = (i%3 == 0, i%5 ==0)
  if mapping[key]:
    print(mapping[key])
  else:
    print(i)

# c) using a one-liner with a ternary expression

for i in range(1, 31):
  print("FizzBuzz" if i%3 == 0 and i%5 == 0 else
        "Fizz" if i%3 == 0 else
        "Buzz" if i%5 == 0 else i)
```

```python
#Write a function describe_model(name, version=1, *tags, **metrics) that prints a summary. Call it three different ways: with only required args,
#with defaults, and with extra keyword metrics like accuracy=0.95, loss=0.12.

def describe_model(name, version=1, *tags, **metrics):
    print(f"Model Name: {name}")
    print(f"Version: {version}")

    # Tags (extra positional arguments)
    if tags:
        print("Tags:", tags)
    else:
        print("Tags: None")

    # Metrics (keyword arguments)
    if metrics:
        print("Metrics:")
        for key, value in metrics.items():
            print(f"  {key}: {value}")
    else:
        print("Metrics: None")

    print("-" * 40)

describe_model("ModelA")
describe_model("ModelB", 2, "nlp", "transformer")
describe_model("ModelC", 3, "vision", "cnn", accuracy=0.95, loss=0.12)
```

```python
#Write a function validate_age(age) that: returns "invalid" if not an int, returns "negative" if < 0, returns "minor" if < 18, else returns "adult".
#Write it first with nested if blocks, then rewrite using early returns (guard clauses). Note which reads better.

#nested if blocks
def validate_age(age):
  if isinstance(age, int):
   if age < 0:
     return "negative"
   elif age < 18:
      return "minor"
   elif age > 18:
      return "adult"

  else:
    return "invalid"

#(guard clauses)
def validateAge(age):
  if isinstance(age, int):
    return "invalid"

  if age < 0:
    return "negative"

  if age < 18:
    return "minor"

  return "adult"

print(validate_age(-8))
print(validate_age(-8.2))
print(validate_age(20))
print(validateAge(-8))
print(validateAge(20))

```

```python
#Prime Number Checker

def check_prime(n):
  for i in range(2, int(n**0.5)+1):
    if n%i == 0:
      return False
  return True

print(f"Is prime number: {check_prime(31)}")
```

```python
#Flexible Calculator (*args)

def calculate_result(operator, *nums):
  if operator == "sum":
     return sum(nums)
  elif operator =="product":
    result = 1
    for num in nums:
      result *= num
    return result

print(f"Addition: {calculate_result('sum', 2,3,4)}")
print(f"Product: {calculate_result('product', 2,3,4)}")
```

```python
# Word Frequency Counter

def word_freq(sentence):
  result = {}
  for word in sentence.split():
    result[word] = result.get(word, 0) + 1
  return result
print(word_freq("apple banana apple orange banana apple"))
```

```python
# Find Top 2 Frequent Elements (Advanced)
def top_two(nums):
  freq = {}
  for i in range (0, len(nums)):
     freq[nums[i]] = freq.get(nums[i], 0) + 1;
  sort_list = sorted(freq.items(), key = lambda item: item[1], reverse=True)
  return [item[0] for item in sort_list[:2]]

print(top_two([1,1,1,2,2,3]))
```

```python
# Read numbers from file → handle errors → return average

def read_numbers(filename):
  nums = []
  try:
    with open(filename, "r") as f:
      for line in f:
        try:
           nums.append(int(line.strip()))
        except ValueError:
           print(f"Invalid number or line {line.strip()}. Skipping this!")

    if not nums:
       return "No valid numbers found in the file!"

    return sum(nums)/len(nums)
  except FileNotFoundError:
    return f"{filename} not found!!"

def write_to_file():
  with open("2.txt", "a") as f:
    f.write("1\n")
    f.write("2\n")

write_to_file()
print(read_numbers("2.txt"))
```

```python
#You have labels = ["cat", "dog", "bird"] and scores = [0.9, 0.7, 0.85]. Use zip() to print pairs. Use enumerate() to print with index. Then combine
#both to print: 0: cat → 0.9. These appear in every ML evaluation loop.

labels = ["cat", "dog", "bird"]
scores = [0.9, 0.7, 0.85]

#zip
for label, score in zip(labels, scores):
    print(label, score)

#enumerate
for i, label in enumerate(labels):
    print(i, label)

#combined
for i, (label, score) in enumerate(zip(labels, scores), start = 1):
    print(f"{i}: {label} → {score}")
```

```python
#Given a paragraph of text (any sentence you like), count the frequency of each word using a dictionary. First write it manually with if key in dict,
# then rewrite using dict.get(key, 0), then again using collections.Counter. Print the top 3 most common words.
from collections import Counter

#using if key in dict,
def count_word_frequency(paragraph):
    freq = {}
    words = paragraph.lower().split()
    for word in words:
      if word in freq:
        freq[word] += 1
      else:
        freq[word] = 1
    sorted_list = sorted(freq.items(), key=lambda item: item[1], reverse=True)
    return sorted_list[:3]

#using get
def count_word_freq(paragraph):
  freq = {}
  words = paragraph.lower().split()
  for word in words:
    freq[word] = freq.get(word, 0) + 1
  sorted_list = sorted(freq.items(), key=lambda item: item[1], reverse=True)
  return sorted_list[:3]

#collections.Counter
def count_words_collections(paragraph):
   frequency = Counter(paragraph.lower().split())
   return frequency.most_common(3)

print("Using if key in dict: ", count_word_frequency("the cat sat on the mat the cat sat"))
print("Using get: ", count_word_freq("the cat sat on the mat the cat sat"))
print("Using collections: ", count_words_collections("the cat sat on the mat the cat sat"))
```

```python
#You have two lists of email addresses (with some duplicates and overlap). Use set operations to find: (a) all unique emails, (b) emails in both
# lists (intersection), (c) emails only in list 1, (d) emails in either but not both (symmetric difference). This pattern is used in train/test
#  split validation.

email_list1 = ["abc@gmail.com", "def@gmail.com", "ABC@gmail.com", "ghi@gmail.com"]
email_list2 = ["jkl@gmail.com", "mno@gmail.com", "ABC@gmail.com", "abc@gmail.com"]

unique_emails = set(email_list1) | set(email_list2)
common_emails = set(email_list1) & set(email_list2)
emails_only_list1 = set(email_list1) - set(email_list2)

print(f"Unique Emails: {unique_emails}")
print(f"Common Emails: {common_emails}")
print(f"Emails only in list 1: {emails_only_list1}")
print(f"Unique but not Common: {set(email_list1) ^ set(email_list2)}")
```

```python
#Create a Dataset class that holds a list of records. Implement: __len__ so len(ds) works, __getitem__ so ds[0] works, __repr__ for a useful print
# output, and an add(record) method. This is the exact interface PyTorch's Dataset class uses.

class Dataset():
  def __init__(self, records = None):
    self.records = records

  def __len__(self):
    print("called length method from this class")
    return len(self.records)

  def __getitem__(self, indx):
    print("called getitem method from this class")
    return self.records[indx]

  def __repr__(self):
    return f"Values from the List is: {self.records} and size is {len(self.records)}"

  def add(self, record):
    self.records.append(record)

ds = Dataset([{'X': 1, 'Y' : 0}, {'X': 2, 'Y' : 1}])
ds.add({'X': 3, 'Y': 2})
print(ds)
print(len(ds))
print(ds[0])

```

```python
class BankAccount:
    def __init__(self, balance):
        self._balance = balance

    def deposit(self, amount):
        self._balance += amount

    def withdraw(self, amount):
        if amount <= self._balance:
            self._balance -= amount

    def show_balance(self):
        return self._balance


acc = BankAccount(1000)

acc.deposit(500)

print(acc.show_balance())
print(acc._balance)

# This works but is not recommended.
# acc._balance = 2000
# print(acc._balance)
```

```python
#Create a Temperature class that stores temp in Celsius internally. Use @property to expose .fahrenheit and .kelvin as computed read-only values.
#Compare this to the Java getter pattern. Try setting temp.fahrenheit = 100 — what happens? Add a setter to handle it.

class Temperature():
  def __init__(self, celsius):
    self._celsius = celsius

  @property
  def celsius(self):
    return self._celsius

  @celsius.setter
  def celsius(self, value):
    self._celsius = value

  @property
  def fahrenheit(self):
    return (self._celsius * 9/5) + 32

  @fahrenheit.setter
  def fahrenheit(self, value):
    self._celsius = (value -32) * 5/9

  @property
  def kelvin(self):
    return self._celsius + 273.15

  def __repr__(self):
    return f"Temperature ({self._celsius:.1f} C)"

temp = Temperature(30)
print(temp.celsius)
temp.fahrenheit = 100 #AttributeError: property 'fahrenheit' of 'Temperature' object has no setter initially, added later
print(temp)


```

```python
#Create a ModelConfig class. Add a @classmethod called from_dict(cls, d) that creates an instance from a dictionary, and a @staticmethod called
#validate_lr(lr) that returns True if learning rate is between 0 and 1. Understand: classmethod receives the class, staticmethod receives nothing.

class ModelConfig():
  def __init__(self, name, lr, epochs):
    self.name = name
    self.lr = lr
    self.epochs = epochs

  def __repr__(self):
    return f'ModelConfig({self.name}, lr = {self.lr}, epochs = {self.epochs})'

  @classmethod
  def from_dict(cls, d):
    return cls(d['name'], d.get("lr", 0.001), d.get("epochs", 10))

  @staticmethod
  def validate_lr(lr):
    return 0 < lr < 1

dict_config = {'name': 'BERT', 'lr': 0.002, 'epochs': 10}
config = ModelConfig.from_dict(dict_config)
print(config)
print("Validate lr:", ModelConfig.validate_lr(dict_config['lr']))
```
