```python
# Recreate this Java class in Python:
# public class BankAccount {
#     private String owner;
#     private double balance;
#     public BankAccount(String owner, double initialBalance) { ... }
#     public void deposit(double amount) { ... }
#     public boolean withdraw(double amount) { ... }
#     public double getBalance() { ... }
#     public String toString() { ... }
# }

# Requirements:
# - deposit(amount): adds to balance, prints confirmation
# - withdraw(amount): deducts if sufficient funds, returns True/False
# - get_balance(): returns current balance
# - __str__: returns "BankAccount[owner=John, balance=$1500.00]"

class BankAccount:

  def __init__(self, owner, initialBalance):
    self.owner = owner
    self.balance = initialBalance

  def deposit(self, amount):
    self.balance += amount
    print(f"Deposited: {amount:.2f}. New Balance: {self.balance:.2f}")

  def withdraw(self, amount):
    if amount < self.balance:
      self.balance -= amount
      print(f"Withdrew : {amount:.2f}. Remaining Balance: {self.balance:.2f}")
      return True
    else:
      print("Insufficient funds!")
      return False

  def get_balance(self):
    return self.balance

  def __str__(self):
    return f"BankAccount[owner={self.owner}, balance=${self.balance:.2f}]"

# Test it:
acc = BankAccount("John", 1000)
acc.deposit(500)
acc.withdraw(200)
acc.withdraw(2000)   # should fail — insufficient funds
print(acc)
```

```python
# Build an Employee hierarchy — similar to what you'd do in Spring Boot entities

# Base class: Employee
#   __init__(name, salary)
#   get_details() → "Employee: John | Salary: $95000"
#   give_raise(percent) → increases salary by percent

# Child class: Manager(Employee)
#   __init__(name, salary, team_size)
#   get_details() → "Manager: John | Salary: $95000 | Team: 5 people"
#   give_raise(percent) → managers get an extra 5% bonus on top

# Child class: Developer(Employee)
#   __init__(name, salary, language)
#   get_details() → "Developer: John | Salary: $95000 | Language: Java"

class Employee:

  def __init__(self, name, salary):
    self.name = name
    self.salary = salary

  def get_details(self):
    return f'Employee: {self.name} | Salary: ${self.salary:,.2f}'

  def give_raise(self, percent):
    self.salary += self.salary * (percent/ 100)

class Manager(Employee):

  def __init__(self, name, salary, team_size):
    super().__init__(name, salary)
    self.team_size = team_size

  def get_details(self):
    return f'Manager: {self.name} | Salary: ${self.salary:,.2f} | Team: {self.team_size} people'

  def give_raise(self, percent):
    super().give_raise(percent + 5)

class Developer(Employee):

  def __init__(self, name, salary, language):
    super().__init__(name, salary)
    self.language = language

  def get_details(self):
    return f'Developer: {self.name} | Salary: ${self.salary:,.2f} | Language: {self.language}'

# Test:
emp = Employee("Alice", 70000)
mgr = Manager("Bob", 95000, 8)
dev = Developer("Charlie", 85000, "Python")
for person in [emp, mgr, dev]:
    print(person.get_details())
    person.give_raise(10)
    print(f"  After raise: ${person.salary:,.2f}")
```

```python
# Here is a Java Spring Boot style Service class.
# Rewrite it completely in Python:

# public class StudentService {
#     private List<Student> students = new ArrayList<>();
#
#     public void addStudent(String name, int age, double gpa) { ... }
#     public Student findByName(String name) { ... }
#     public List<Student> getHonorRoll() { return gpa >= 3.5 }
#     public double getClassAverage() { ... }
#     public void printRoster() { ... }
# }
#
# public class Student {
#     String name; int age; double gpa;
# }

class Student:
  def __init__(self, name, age, gpa):
    self.name = name
    self.age = age
    self.gpa = gpa

  def __str__(self):
    return f'{self.name} {self.age} {self.gpa}'

class StudentService():
  def __init__(self):
    self.students = []

  def add_student(self, name, age, gpa):
    print(Student(name, age, gpa))
    self.students.append(Student(name, age, gpa))

  def findByName(self, name):
    for s in self.students:
      if s.name.lower() == name.lower():
        return f"{name} name found!"

      return "None"

  def getHonorRoll(self):
    return [s for s in self.students if s.gpa >=3.5]

  def getClassAverage(self):
    return sum(s.gpa for s in self.students)/len(self.students)

  def printRoster(self):
    print(f"{'Name':<15} {'Age':>4} {'GPA': >6}")
    print("-" * 28)
    for s in self.students:
      print(f"{s.name:<15} {s.age:>4} {s.gpa:>6.2f}")


# Test with at least 5 students, print honor roll and class average
svc = StudentService()
svc.add_student("Alice",20, 3.5)
svc.add_student("Bob",30, 3.4)
svc.add_student("John",25, 3.2)
svc.add_student("Bindu",19, 3.8)
svc.add_student("Don",29, 3.0)
print(svc.findByName("Alice"))
print(f"Honor Roll: {[s.name for s in svc.getHonorRoll()]}")
print(f"Class Average GPA :  {svc.getClassAverage():.2f}")
svc.printRoster()
```
