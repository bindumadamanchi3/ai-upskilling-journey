```python
# Inspect a DataFrame (Easy)
#Using the engineers df, print: shape, column names, data types, first 3 rows, last 3 rows, and a statistical summary of numeric columns. Use 6 different Pandas methods.

import pandas as pd

df = pd.read_csv(io.StringIO(csv_data))
print(df)

print(df.shape)
print(df.columns.tolist())
print(df.dtypes)
print(df.head(3))
print(df.tail(3))
print(df.describe())
```

```python
#Add and drop columns (Easy)
# 3a. Add salary_monthly (salary / 12, rounded to 2 decimal places).
# 3b. Add seniority column — 'Senior' if experience_years >= 7, else 'Junior'.
# 3c. Drop the passed_screening column.

import numpy as np

print(df.to_string)

df['salary_monthly'] = (df['salary']/12).round(2)

df['seniority'] = np.where(df['experience_years'] >= 7, 'Senior', 'Junior')

df = df.drop(columns = ['passed_screening'])
print(df.to_string())
```

```python
#Handle missing values after loading
# 7a. Count nulls per column. 7b. Fill missing salaries with the column median. 7c. Fill missing cities with 'Unknown'. 7d. Drop any rows still with nulls.

dirty_csv = '''name,salary,city
Alice,95000,New York
Bob,,San Francisco
Carol,105000,
Dave,,
Eve,98000,Seattle'''

df_dirty = pd.read_csv(io.StringIO(dirty_csv))

print(df_dirty.info())
print(df_dirty.isnull().sum())

median_sal = df_dirty['salary'].median()   # 98000.0
df_dirty.fillna({'salary': median_sal}, inplace = True)
#OR
df_dirty['salary'] = df_dirty['salary'].fillna(median_sal)

df_dirty['city'] = df_dirty['city'].fillna("Unknown")

print(df_dirty.dropna())
print(df_dirty.isnull().sum())
```

```python
# 10a. Engineers in Austin OR Seattle.
# 10b. Engineers with salary > 100k AND experience < 6 years.
# 10c. Engineers NOT in San Francisco.

df_cities = df[df['city'].isin(['Austin', 'Seattle'])]
print(df_cities[['name', 'city']])

df_salary_exp = df[(df['salary'] > 100000) & (df['experience_years'] < 6)]
print(df_salary_exp[['name', 'city']])

df_not_sf = df[~(df['city'] == 'San Francisco')]
print(df_not_sf[['name', 'city']])
```

```python
#Sort filtered results (Easy)
# 13a. Engineers who passed screening, sorted by salary descending.
# 13b. All engineers sorted by city (A-Z) then salary (descending) within each city.
# 13c. Top 3 highest-paid engineers.

d = df[df['passed_screening'] == True].sort_values('salary', ascending=False)
print(d[['name', 'city']])

e = df.sort_values(['city','salary'], ascending = [True, False])
print(e[['name', 'city']])

f = df.sort_values(['salary'], ascending = False)
print(f[['name', 'city', 'salary']].head(3))
#OR
g = df.nlargest(3, 'salary')
print(g[['name', 'city', 'salary']])
```

```python
#Group and aggregate after filtering (Medium)
# 14a. Average salary per city across all engineers.
# 14b. Count of engineers with salary > 95000 per city.
# 14c. Which city has the highest average salary?

avg_by_city = df.groupby('city')['salary'].mean().round(0)
print(avg_by_city)

high_sal = df[df['salary'] > 95000]
print(high_sal.groupby('city').size())

high_avg = avg_by_city.sort_values(ascending = True)
print(high_avg.tail(1))
#OR
print(avg_by_city.idxmax()) #idxmax returns the label of the row with the highest value. whereas max() returns the highest value/number
```

```python
# COMBINED CHALLENGE
# Step 1. Load the CSV and check for nulls.
# Step 2. Add salary_monthly, seniority (Senior if years >= 7), salary_band (High if >= 110k).
# Step 3. Filter to engineers who are Senior AND passed screening AND salary_band is High.
# Step 4. Group that filtered set by city — show count, average salary, average experience.
# Step 5. Print the city where the most qualifying engineers are based.

import numpy as np

#print(df)

df['salary_monthly'] = (df['salary']/12).round(2)
df['seniority'] = np.where(df['experience_years'] >= 7, 'Senior', 'Junior')
df['salary_band'] = np.where(df['salary'] >= 110000, 'High', 'Standard')
print(df)

filter_eng = df.query('seniority == "Senior" and passed_screening == True and salary_band == "High"')
group_filtered = filter_eng.groupby('city').agg({
    'city': 'size',
    'salary': 'mean',
    'experience_years': 'mean'
})
print(group_filtered)

print(group_filtered['city'].idxmax())
```
