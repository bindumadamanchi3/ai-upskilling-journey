```python
# 1. Write a list of 5 programming languages to "languages.txt"
#    (one per line)
# 2. Read the file back and print each line numbered:
#    1. Python
#    2. Java
#    3. JavaScript
#    ...
# 3. Append 2 more languages to the same file without overwriting
# 4. Read and print the final file
languages =['C', 'C++', 'Java', 'JavaScript', 'Python']

try:
  with open("languages.txt", "w") as file:
    for lang in languages:
      file.write(lang + '\n')


  with open("languages.txt", "r") as file:
    for i, line in enumerate(file, start=1):
      print(f"{i}. {line.strip()}")

  with open("languages.txt", "a") as file:
    file.write('C#\n')
    file.write('.NET\n')

  with open("languages.txt", "r") as file:
    print(file.read())

except FileNotFoundError:
  print("File not Found!!")

finally:
  print("Execution completed")
```

```python
# Given this data, write it to "employees.csv" then read it back
import csv

employees = [
    {"name": "Alice",   "role": "Engineer",   "salary": 90000},
    {"name": "Bob",     "role": "Manager",    "salary": 110000},
    {"name": "Charlie", "role": "Designer",   "salary": 80000},
    {"name": "Diana",   "role": "Engineer",   "salary": 95000},
]

try:
  #write to file
  with open("employees.csv", "w", newline = "") as f:
    writer = csv.DictWriter(f, fieldnames= ["name", "role", "salary"])
    writer.writeheader()
    writer.writerows(employees)

  #read it back
  with open("employees.csv", "r") as f:
    reader = csv.DictReader(f)
    rows = list(reader)
    for row in rows:
      print(dict(row))

  #Filter and print only engineers
  engineers = [r for r in rows if r['role'] == "Engineer"]
  print('Engineers ', [e['name'] for e in engineers])

  #Calculate average salary from the CSV data
  average_salary = sum(int(r['salary']) for r in rows)/ len(rows)
  print(f"Average Salary : {average_salary:.2f}")

except FileNotFoundError:
  print("File not Found!!")
except ValueError:
  print("Invalid Numbers!")
```

```python
# Build a contact book that saves to and loads from a file

# Features:
# 1. Load contacts from "contacts.csv" on startup (handle if file doesn't exist)
# 2. add_contact(name, phone, email) — adds and immediately saves to file
# 3. find_contact(name) — case-insensitive search, returns contact or None
# 4. delete_contact(name) — removes and saves
# 5. list_all() — prints all contacts formatted
# 6. On exit, print "X contacts saved."
import csv

class ContactBook():
  def __init__(self, filename="contacts.csv"):
    self.filename = filename
    self.contacts = []
    self.load_contacts_startup()

  def load_contacts_startup(self):
    try:
      with open(self.filename, "r", newline="") as file:
        reader = csv.DictReader(file)
        self.contacts = list(reader)

    except FileNotFoundError:
      self.contacts = []
      print("File not found! Starting with empty contact list.")

  def save_contacts(self):
    with open(self.filename, "w", newline="") as file:
      writer = csv.DictWriter(file, fieldnames= ["name", "phone", "email"])
      writer.writeheader()
      writer.writerows(self.contacts)

  def add_contact(self, name, phone, email):
    contact = {"name": name, "phone": phone, "email": email}
    self.contacts.append(contact)
    self.save_contacts()
    print("Contact successfully added")
    print(self.contacts)

  def find_contact(self, name):
    for contact in self.contacts:
      if contact['name'].lower() == name.lower():
        print(f"{name} found in Contact Book!")
        return contact
    print(f"{name} not found in Contact Book!")
    return None

  def delete_contact(self, name):
    for contact in self.contacts:
      if contact['name'].lower() == name.lower():
        self.contacts.remove(contact)
        self.save_contacts()
        print(f"Contact '{name}' delete Successful!")
        return;
    print(f"{name} not found in Contact Book!")

  def list_all(self):
    if not self.contacts:
      print("No contacts available!")
      return

    for i,contact in enumerate(self.contacts, start=1):
      print(f"{i}. {contact['name']} | {contact['phone']} | {contact['email']}")

  def close(self):
    print(f"{len(self.contacts)} contacts saved.")


# Test: add 4 contacts, find one, delete one, list all, exit
cb = ContactBook()
cb.add_contact("Alice", "8162996374", "alice@gmail.com")
cb.add_contact("John", "8162996370", "john@gmail.com")
cb.add_contact("Bob", "8162926374", "bob@gmail.com")
cb.add_contact("Kenny", "8162993374", "kenny@gmail.com")
cb.find_contact("john")
cb.find_contact("abc") #should return name not found
cb.delete_contact("bob")
cb.delete_contact("bob") #should return contact not found
cb.list_all()
cb.close()



```
